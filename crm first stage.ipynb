```python
import pandas as pd
import sqlite3
```

```python
file_path = "accounts.csv"  # path to the CSV
df1 = pd.read_csv(file_path)
file_path = "products.csv"  # path to the CSV
df2 = pd.read_csv(file_path)
file_path = "sales_pipeline.csv"  # path to the CSV
df3 = pd.read_csv(file_path)
file_path = "sales_teams.csv"  # path to the CSV
df4 = pd.read_csv(file_path)
```

```python
db_path = "my_database.db"  # SQLite database file
conn = sqlite3.connect(db_path)
```

```python
from sqlalchemy import create_engine

engine = create_engine(f"sqlite:///{db_path}")
```

```python
table_name = "Accounts"  # choose a table name
df1.to_sql(table_name, conn, if_exists="replace", index=False)
table_name = "Products"  # choose a table name
df2.to_sql(table_name, conn, if_exists="replace", index=False)
table_name = "sales_pipelines"  # choose a table name
df3.to_sql(table_name, conn, if_exists="replace", index=False)
table_name = "sales_teams"  # choose a table name
df4.to_sql(table_name, conn, if_exists="replace", index=False)
tables = pd.read_sql("SELECT name FROM sqlite_master WHERE type='table';", conn)
print("Tables:", tables['name'].tolist())
```
